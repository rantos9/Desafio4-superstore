This notebook's purpose is to generate pickle files for the fbprophet models

```python
import pandas as pd
from fbprophet import Prophet
```

```python
df = pd.read_excel('../data/Sample - Superstore.xls')
```

```python
#Preparando função de preparação da Time Series
def prepare_time_series (data, category):
    data = data.loc[df['Category'] == category]
    cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name',
            'Segment', 'Country', 'City','State', 'Postal Code','Region', 'Product ID', 'Category',
            'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']
    data.drop(cols, axis=1, inplace=True)
    data = data.sort_values('Order Date')
    data = data.groupby('Order Date')['Sales'].sum().reset_index()
    data = data.set_index('Order Date')
    data = data['Sales'].resample('MS').mean()
    data = data.reset_index()
    data = data.rename(columns= { "Order Date":"ds" , "Sales": "y" })
    return data
```

```python
def fitModel(data):
    model = Prophet()
    model.fit(data)
    return model
```

```python
def predict(model, periods = 6):
    future_dates = model.make_future_dataframe(periods=periods, freq='MS')
    forecast = model.predict(future_dates)
    return forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
def savePickle(results, category):
    import pickle
    pickle.dump(results, open(category+".pkl","wb"))
    print('Pickle file saved as '+category+".pkl")
```

```python
for x in df.Category.unique():
    category = x
    data = prepare_time_series(df, category)
    model = fitModel(data)
    savePickle(model, category)
```
